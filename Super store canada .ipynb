```python
df1=df.drop(['CustID','PostalCode','City', 'Country','Market','ProdID', 'RefID', 'CustName'], axis=1)
df1
```

```python
df1.columns
```

```python
#Create a new column total sales
df1['TotalSales']= df1['SalePrice']*df1['Quantity']
df1.head(3)
```

```python
#Create a new column total cost
df1['TotalCost']= df1['CostPrice']*df1['Quantity']
df1.head(3)
```

```python
df1['Discount'].unique()
```

```python
#Calculating discount amount
df1['discount_perc'] = df1['Discount']*100
df1.head(3)
```

```python
#Calculating the discount amount
df1['Discount_amount'] = df1['discount_perc']*df1['TotalSales']/100
df1.head(2)
```

```python
#Sales after Discount
df1['GrossSales']= df1['TotalSales']- df1['Discount_amount']
df1.head(2)
```

```python
#Calculating Cost Price
df1['GrossCostPrice'] = df1['TotalCost']+df1['ShipCost']
df1.head(2)
```

```python
#Create a column for Profit
df1['Profit']= df1['GrossSales'] - df1['GrossCostPrice']
df1.head(3)
```

### Q1 What is the % of Sales of Appliances in the Year 2012?

```python
#extracting year from order date
df1['Year'] = pd.DatetimeIndex(df1['OrderDate']).year
df1.head(3)
```

```python
year_wise=pd.DataFrame(df1.groupby('Year').sum()['TotalSales'])
year_wise
```

```python
App_2012 = df1[(df1.ProdSubCat=='Appliances') & (df1.Year==2012) & (df1.Status=='OK')]
```

```python
App_sales = App_2012['TotalSales'].sum()
App_sales
```

```python
Sales_2012 = df1[(df1.Year==2012) & (df1.Status=='OK')]
Sales_year_2012 = Sales_2012['TotalSales'].sum()
Sales_year_2012
```

```python
print("% of application on sales in 2012 (in %):", (App_sales*100)/Sales_year_2012)
```

### Q2 What is the % of Sales of “Home Office” in the Year 2011?

```python
HO_2011 = df1[(df1.CustomerSegment=='Home Office') & (df1.Year==2011) & (df1.Status=='OK')]
```

```python
HO_Sales = HO_2011['TotalSales'].sum()
HO_Sales
```

```python
Sales_2011 = df1[(df1.Year==2011) & (df1.Status=='OK')]
Sales_year_2011 = Sales_2011['TotalSales'].sum()
Sales_year_2011
```

```python
print("% of Home Office on sales in 2011 (in %):", (HO_Sales*100)/Sales_year_2011)
```
